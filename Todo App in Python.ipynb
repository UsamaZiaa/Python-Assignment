```python
#To Do List in Python!

def isInteger(userInput):
    try:
        int(userInput)
        return True
    except ValueError:
        return False 

def mainMenu(menu):
    print("\n---------------\n|  Main Menu  |\n---------------\n")
    for (mainNumber, mainName) in menu.items():
        print("[{}] - {}".format(mainNumber, mainName))

    userInput = input("\n[-] Enter Command ~ ") 
    return userInput
    
def addTasks(tasksDictionary):
    print("\n--------------------\n|   Adding Tasks   |\n--------------------")
    while(True):
        taskIndex = len(tasksDictionary) + 1
        
        if taskIndex > 1:
            taskIndex = int(max(tasksDictionary.keys())) + 1
        
        task = input("\n[-] Task #{} ~ ".format(taskIndex))
        if task.strip() == "":
            print("[-] Empty Task Not Allowed!\n")
            continue

        tasksDictionary.update({taskIndex : task})
        print("[-] Task #{} Added Successfully\n".format(taskIndex))
        userInput = input("[-] Do You Want to Add Another Task? [Y/N] ~ ")

        if userInput.lower() == "y":
            continue
        else:
            break

def deleteTasks(tasksDictionary):
    print("\n----------------------\n|   Deleting Tasks   |\n----------------------\n")
    displayTasks = True
    while(True):
        if len(tasksDictionary) > 0:
            
            if displayTasks:
                for (taskIndex, task) in tasksDictionary.items():
                    print("[Task #{}] - {}".format(taskIndex, task))                
                displayTasks = False

            taskIndexInput = input("\n[-] Enter Task Index (eg. 1) ~ ")
            
            if isInteger(taskIndexInput):
                taskIndex = int(taskIndexInput)
                if taskIndex in tasksDictionary.keys():
                    tasksDictionary.pop(taskIndex)
                    print("[-] Task #{} Deleted Successfully\n".format(taskIndex))
                    userInput = input("[-] Do You Want to Delete Another Task? [Y/N] ~ ")
                else:
                    print("[-] Invalid Task Index\n[-] Try Valid Task Index!\n")
                    continue

                if userInput.lower() == "y":
                    continue
                else:
                    break  

            else:
                print("[-] Invalid Task Index\n[-] Try Valid Task Index!\n")
                continue
        else:
            print("[-] No Tasks Found!\n")
            userInput = input("[-] Do You Want to Add Task? [Y/N] ~ ")

            if userInput.lower() == "y":
                addTasks(tasksDictionary)
                break
            else:
                break
            
def viewTasks(tasksDictionary):
    print("\n-------------------\n|  Viewing Tasks  |\n-------------------\n")
    if len(tasksDictionary) > 0:
        for (taskIndex, task) in tasksDictionary.items():
            print("[Task #{}] - {}".format(taskIndex, task))
    else:
        print("[-] No Task Found!")
    
    userInput = input("\n[-] Do You Want to Add Tasks? [Y/N] ~ ")
    
    if userInput.lower() == 'y':
        return 1
    else:
        return 0

def aboutDeveloper():
    print("\n---------------------\n|  About Developer  |\n---------------------\n")
    print("""
                    My name is Usama Zia, I am Passionate Python Programmer! 
    """)
    
menu = {
    1 : "Add Tasks",
    2 : "View Tasks",
    3 : "Delete Tasks",
    4 : "Terminate Program",
    5 : "About Developer" 
}

tasksDictionary = {}
print("----------------------------\n|        Things to Do      |\n----------------------------")

while(True):

    userCommand = mainMenu(menu)
    
    if isInteger(userCommand):
        userCommand = int(userCommand)
        
        # Adding Tasks
        if userCommand == 1:
            addTasks(tasksDictionary)
            continue
            
        # Viewing Tasks
        if userCommand == 2:
            addTask = viewTasks(tasksDictionary)
            if addTask == 1:
                addTasks(tasksDictionary)
            else:
                continue
            
        # Deleting Tasks
        if userCommand == 3:
            deleteTasks(tasksDictionary)
            
        # Terminating Program
        if userCommand == 4:
            print("[+] Program Terminated")
            break
        
        # About Developer
        if userCommand == 5:
            aboutDeveloper()
            
    else:
        print("[-] Invalid Command\n[-] Try Valid Command Between {} to {}\n".format(1, len(menu)))
        
```
